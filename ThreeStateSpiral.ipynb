```python
from matplotlib.animation import FuncAnimation
from IPython import display
import numpy as np
import matplotlib.pyplot as plt
```

```python
GridSize = 100
Automata = np.zeros((GridSize,GridSize))
MaxIterations = 10

# Setting initial condition
Automata[ int(GridSize/3):int(2*GridSize/3) , int(GridSize/2) ]   = 2
Automata[ int(GridSize/3):int(2*GridSize/3)  , int(GridSize/2)+1 ] = 1

fig = plt.figure()

I = plt.imshow(Automata)

def animate(frame):

    global Automata
    
    NextAutomata = np.copy(Automata)
    
    for i in range(1,GridSize-1):
        for j in range(1,GridSize-1):    
    
            if Automata[i,j]==0 :
                for k in range(3):
                    for l in range(3):
                        if Automata[i+k-1,j+l-1]==2 :
                            NextAutomata[i,j]=2
                            break
                        
            elif Automata[i,j]==1 :
                NextAutomata[i,j]=0
            
            elif Automata[i,j]==2 :
                NextAutomata[i,j]=1
              
    Automata = NextAutomata
    I.set_data(NextAutomata)
                   
anim = FuncAnimation(fig, animate, frames= 100, interval=100)
video = anim.to_html5_video()
html = display.HTML(video)
display.display(html)
plt.close()
```
